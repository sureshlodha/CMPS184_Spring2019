```python
# Import web-scraped data

import pandas as pd
import re

#df = pd.read_csv('trulia2_latlon2.csv')

df = pd.read_csv('data/trulia_lat_lon1.csv')



print(df['Features'].head(5))
```

```python
# Used RegEx to parse features

features_list = df['Features']

beds_list = []
baths_list = []
sqft_list = []
#lotsize_list = []
price_per_sqft_list = []
built_in_list = []
heating = []
#rooms_list = []


for ind, features in enumerate(features_list):
    SQFT = False
    beds = '?'
    baths = '?'
    sqft = '?'
    lotsize = '?'
    price_per_sqft = '?'
    built_in = '?'
    rooms = '?'

    for f in features.split(';'):
        if re.match('\d* (Beds)$', f):
            beds=f
        elif re.match('(\d+(.)?\d|\d+) (Baths)$',f):
            baths=f
        elif re.match('^(\d+%\d+|\d+) (sqft)$',f):
            sqft = f
        #elif re.match('^(Lot Size: )(\d+%\d+ sqft|\d+ sqft|\d+ acres|\d+\.\d+ acres)',f):
            #lotsize = f
        elif re.match('(\$\d\d\d\/)(sqft)',f):
            price_per_sqft = f
        elif re.match('(Built in) (\d\d\d\d)',f):
            built_in = f
        #elif re.match('(\d) (Rooms)', f):
            #rooms = f
            

    beds_list.append(beds)
    baths_list.append(baths)
    sqft_list.append(sqft)
    #lotsize_list.append(lotsize)
    price_per_sqft_list.append(price_per_sqft)
    built_in_list.append(built_in)
    #rooms_list.append(rooms)

    
    

df['Beds'] = beds_list
df['Baths'] = baths_list
df['Sqft'] = sqft_list
#df['Lot Size'] = lotsize_list
df['Price_Per_Sqft'] = price_per_sqft_list
df['Year Built'] = built_in_list
#df['Rooms'] = rooms_list

#df.to_csv('Trulia_Lat_Lon_2_Feat.csv')

df.to_csv('data/Trulia_Lat_Lon_Feat.csv')
```

```python
#data = pd.read_csv('Trulia_Lat_Lon_2_Feat.csv')

data = pd.read_csv('data/Trulia_Lat_Lon_Feat.csv')



```

```python
data.head()
```

```python
# Sort out rows that dont contain any ?'s in Lat, Lon, Beds, Baths, Sqft, Price_Per_Sqft, Year Built

data = data.loc[data['Lat'] != "?"]
data = data.loc[data['Lon'] != "?"]
data = data.loc[data['Beds'] != "?"]
data = data.loc[data['Baths'] != "?"]
data = data.loc[data['Sqft'] != '?']
data = data.loc[data['Price_Per_Sqft'] != "?"]
data = data.loc[data['Year Built'] != "?"]
#data = data.loc[data['Rooms'] != '?']
```

```python
# Rows & Columns after filtering
data.shape
```

```python
# Discard irrelevant columns 
data.head()
```

```python
data.columns

data = data.drop(['Unnamed: 0', 'Unnamed: 0.1', 'State', 'State_Abr','County', 'Features','History Events', 'Property Events'], axis = 1)

#data = data.drop(['Unnamed: 0', 'Unnamed: 0.1','Unnamed: 0.1.1', 'State', 'State_Abr','County', 'Features','History Events', 'Property Events'], axis = 1)
```

```python
data.head()
```

```python
# Adjusting columns for Modeling


data['Trulia Est'] = data["Trulia Est"].str.replace("%", "")
data['Trulia Est'] = data["Trulia Est"].str.replace("$", "")

data['Sqft'] = data['Sqft'].str.replace("%", "")
data['Sqft'] = data['Sqft'].str.replace("sqft", "")

data['Beds'] = data['Beds'].str.replace("Beds", "")
data['Baths'] = data['Baths'].str.replace("Baths", "")

data['Price_Per_Sqft'] = data['Price_Per_Sqft'].str.replace('$', "")
data['Price_Per_Sqft'] = data['Price_Per_Sqft'].str.replace('/sqft', "")

data['Year Built'] = data['Year Built'].str.replace("Built in", "")
```

```python
data.dtypes
```

```python
data.isnull().values.any()
```

```python
# Preprocessing Data for buidling prediction model

# Use LabelEncoder() to convert all string labels into numeric

#City              object
#Street            object
#Lat               object
#Lon               object
#Trulia Est        object
#Beds              object
#Baths             object
#Sqft              object
#Price_Per_Sqft    object
#Year Built  object

import numpy as np

data["Lat"] = data["Lat"].astype(np.float64)
data["Lon"] = data["Lon"].astype(np.float64)
data["Trulia Est"] = data["Trulia Est"].astype(np.float64)
data["Beds"] = data["Beds"].astype(np.float64)
data["Baths"] = data["Baths"].astype(np.float64)
data["Sqft"] = data["Sqft"].astype(np.float64)
data["Price_Per_Sqft"] = data["Price_Per_Sqft"].astype(np.float64)
data["Year Built"] = data["Year Built"].astype(np.float64)


from sklearn import preprocessing 

le = preprocessing.LabelEncoder()

data['City'] = le.fit_transform(data['City'])
data['Street'] = le.fit_transform(data['Street'])
```

```python
data.dtypes
```

```python
data.head()
```

```python
from sklearn.cluster import KMeans
import numpy as np

from sklearn.metrics import accuracy_score

kmeans = KMeans(n_clusters = 6, random_state = 0).fit(data)

kmeans.labels_

data['Cluster'] = kmeans.labels_ 
```
