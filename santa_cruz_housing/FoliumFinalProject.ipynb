```python
# df = pd.read_csv('Trulia_Lat_Lon_Feat_with_cluster - Trulia_Lat_Lon_Feat_with_cluster.csv')
# df = pd.read_csv('Trulia_Lat_Lon_Feat_with_10_cluster - Trulia_Lat_Lon_Feat_with_10_cluster.csv')
#read in this csv file with all the features and cluster number
# df = pd.read_csv('Trulia_Lat_Lon_Feat_with_6_cluster - Trulia_Lat_Lon_Feat_with_6_cluster.csv')
df = pd.read_csv('data/95003_Santa_Cruz_with_6_clusters.csv')


#gets the numbers from column names Lat and Lon and finds the mean which is able to have a starting point for the map
map0 = folium.Map(location=[df['Lat'].mean(), 
 df['Lon'].mean()], 
 zoom_start=12)

mc = MarkerCluster()
#creating a Marker for each point in df_sample. Each point will get a popup with their zip
for row in df.itertuples():
    #this checks for the cluster number and compares the current cluster number to 0-5 
    if row.Cluster == 0:
        mc.add_child(folium.Marker(location=[row.Lat,  row.Lon], popup = ('Street:', row.Street, 'Beds:', 
                                                                          row.Beds, 'Baths:', row.Baths, 
                                                                          'Sqft:', row.Sqft,'Price_Per_Sqft:', 
                                                                          row.Price_Per_Sqft, 'Year Built:', 
                                                                          row.YearBuilt, 'Trulia Est:', 
                                                                          row.TruliaEst)))
        map0.add_child(mc)

#this saves the map into an html file it can properly render the markers
map0.save('folium/index0.html')
```
